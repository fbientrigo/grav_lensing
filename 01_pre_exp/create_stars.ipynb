```python
import numpy as np
import matplotlib.pyplot as plt

N = 500
x,y = np.random.randint(0, 100, N), np.random.randint(0, 100, N)

# Create the scatter plot
plt.scatter(x, y, color='black')

# Remove the axes
plt.axis('off')

# Save the image
plt.savefig('scatter_plot.png', bbox_inches='tight', pad_inches=0)

# Show the plot
plt.show()
```
